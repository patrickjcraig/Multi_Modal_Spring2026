```python
import open3d as o3d
```

```python
# Import our file TestPart as a 3D mesh.
mesh = o3d.io.read_triangle_mesh("TestPart.stl")
mesh.compute_vertex_normals()
```

```python
# Cast our TestPart mesh to a point cloud (pcd)
pcd = mesh.sample_points_poisson_disk(number_of_points=2000)
```

```python
# For Jupyter notebooks, use the web visualizer
# This will open in a new browser tab/window
o3d.web_visualizer.draw([{"name": "mesh", "geometry": mesh}, 
                          {"name": "point_cloud", "geometry": pcd}])
```

```python
# Alternative: Extract points for custom visualization with plotly
import numpy as np
import plotly.graph_objects as go

# Get point cloud points and colors (if any)
points = np.asarray(pcd.points)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=points[:, 0],
    y=points[:, 1],
    z=points[:, 2],
    mode='markers',
    marker=dict(size=2, color=points[:, 2], colorscale='Viridis')
)])

fig.update_layout(
    scene=dict(aspectmode='data'),
    title='Point Cloud Visualization'
)
fig.show()
```
